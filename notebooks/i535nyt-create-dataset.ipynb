```python
def tokenize(sentence):
    words = re.sub('\.|"',' ', sentence).split()
    result = []
    for word in words:
        w = re.sub(r'[^a-zA-Z0-9]+', '', word).split()
        if isinstance(w, list):
            for i in w:
                result.append(i.lower())
        else:
            result.append(w.lower())
    return result
```

```python
DIR = "../input/nyt-articles-data"
FILE = DIR + '/df_{}.csv'
```

```python
for year in range(1921, 2021, 10):
    df = pd.concat([pd.read_csv(FILE.format(i + year)) for i in trange(10)])
    df = df[['year', 'sentence']].reset_index(drop=True)
    df.sentence = df.sentence.swifter.apply(tokenize)
    print(year, df.shape)
    df.to_csv("df_{}_to_{}.csv".format(year, year + 9), index=False)
    del df
    gc.collect()
    print(year)
```
